# Reduce dataset

```python
import pandas as pd
from pathlib import Path


data = pd.read_csv(Path("../data/Combined Data.csv"))

data.head()
```

```python
data.columns.to_list()
```

```python
import sys


def get_size(df: pd.DataFrame) -> str:
    file_size = sys.getsizeof(df)
    return f"File size: {(file_size / 2048):,.2f} mb"


get_size(data)
```

```python
sample_data = data.sample(frac=0.05)

print(get_size(sample_data))
```

```python
sample_data.to_csv(Path("../data/sample_data.csv"), index=False)
```
